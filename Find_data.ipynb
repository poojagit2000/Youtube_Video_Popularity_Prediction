# Import Required Libraries

```python
# import googleapiclient to fetch data from youtube api
import pandas as pd
import numpy as np
import getpass
from googleapiclient.discovery import build
```

# Get data from kaggle

```python
# Get data for the countries US,CA,IN,AU
raw_data_us = pd.read_csv('TrendingYT_VideosUS.csv')
raw_data_mixed = pd.read_csv('TrendingYT_VideosMixed.csv')
raw_data = pd.concat([raw_data_us,raw_data_mixed],axis=0)
raw_data
```

```python
raw_data.columns
```

# Drop Duplicates in the Video Id

```python
raw_data = raw_data.drop_duplicates(subset=['video_id'], keep='first')
```

```python
raw_data.info()
```

```python
video_id_list = list(raw_data['video_id'])
video_id_list
```

```python
# Set up the YouTube API
# Prompt the user for a apikey without echoing
api_key = getpass.getpass(prompt='Enter your api_key: ')
#api_key = 'AIzaSyAa3Xi7u3Dk1P0hU6B9PO9d64jl43h3zJs'
youtube = build('youtube', 'v3', developerKey=api_key)

# Function to get video details
def get_video_details(video_ids):
    video_details = []

    # Make multiple requests with a maximum of 50 video IDs per request
    for i in range(0, len(video_ids), 50):
        batch_ids = video_ids[i:i + 50]
        request = youtube.videos().list(
            part="snippet,contentDetails,statistics,status",
            id=",".join(batch_ids)
        )
        response = request.execute()

        for video in response['items']:
            details = {
                'video_id': video['id'],
                'channel_title': video['snippet']['channelTitle'],
                'duration': video['contentDetails']['duration'],
                'dislike_count': video['statistics'].get('dislikeCount', 0),
                'category_id': video['snippet']['categoryId'],
                'favorite_count': video['statistics'].get('favoriteCount', 0),
                'dimension': video['contentDetails']['dimension'],
                'definition': video['contentDetails']['definition'],
                'caption_status': video['contentDetails']['caption'],
                'projection': video['contentDetails']['projection'],
                'live_broadcast_content': video['snippet']['liveBroadcastContent'],
                'default_audio_language': video['snippet'].get('defaultAudioLanguage', 'N/A')
            }
            video_details.append(details)

    return video_details

# List of video IDs
video_ids = list(raw_data['video_id'])
# Get video details
video_details = get_video_details(video_ids)

# Convert to DataFrame
youtube_video_details = pd.DataFrame(video_details)

# Display the DataFrame
youtube_video_details.info()
```

# Merge kaggle data using youtube API

```python
youtube = pd.merge(raw_data, youtube_video_details, on='video_id', how='outer')
youtube.info()
```

# Save the DataFrame to CSV format

```python
youtube.to_csv('youtube.csv',index=False)
```
